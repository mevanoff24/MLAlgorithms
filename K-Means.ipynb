# K-Means

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from clustering.kmeans import KMeans
from utils import *
```

```python
X, true_clusters = get_clustering_data(random_state=100, cluster_std=3.0)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=true_clusters, s=25, edgecolor='k');
```

### Fit Model

```python
kmeans = KMeans(K=3, max_iter=3)
preds = kmeans.fit(X)
```

### Plot results

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=true_clusters, s=25, edgecolor='k')
plt.title('True Clusters')
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=preds, s=25, edgecolor='k')
plt.title('K-Means Clusters');
```
