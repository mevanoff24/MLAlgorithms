# Logistic Regression

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from linear_model.logistic_regression import LogisticRegression
from utils import *
```

```python
X, y = get_simple_classification_data(random_state=104)
# train / test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=50, random_state=100)
# visualize dataset
plt.scatter(X_train[:, 0], X_train[:, 1], marker='o', c=y_train, s=25, edgecolor='k')
plt.show()
```

```python
# init model
clf = LogisticRegression()
# fit model 
clf.fit(X_train, y_train)

train_accuracy = accuracy(y_train, clf.predict(X_train, return_probs=False))
test_accuracy = accuracy(y_test, clf.predict(X_test, return_probs=False))

print('train accuracy', train_accuracy)
print('test accuracy', test_accuracy)
```

```python
# plot Decision Boundary
plt.scatter(X_train[:, 0], X_train[:, 1], marker='o', c=y_train, s=25, edgecolor='k')
X_values = [np.min(X_train[:, 0] - 1), np.max(X[:, 1] + 1)]
y_values = -(clf.b + np.dot(clf.W[0], X_values)) / clf.W[1]

plt.plot(X_values, y_values, label='Decision Boundary')
plt.legend()
plt.show()
```

# Pros / Cons

- The big thing that should be noted here is that a Logistic Regression model is a linear classifier [[1]](http://stats.stackexchange.com/questions/79259/how-can-i-account-for-a-nonlinear-variable-in-a-logistic-regression) [[2]](http://stats.stackexchange.com/questions/93569/why-is-logistic-regression-a-linear-classifier). It cannot fit non-linear data. Thus, the model creates a single straight line boundary between the classes.
- Interpretable by examining the odd-ratio for the coefficients
- Great for probabilities, since this works based on the sigmoid function
- Can set a threshold value
- Supports [online learning](https://vision.cornell.edu/se3/wp-content/uploads/2014/09/osb_iccv09_cam.pdf)
